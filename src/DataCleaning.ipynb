clean data

```python
# Set True only once to clean the dataset
CLEAN_DATASET = True

# Keep track of removed images
removed_files = []

if CLEAN_DATASET:
    for image_class in os.listdir(data_dir):
        class_path = os.path.join(data_dir, image_class)
        if not os.path.isdir(class_path):
            continue

        for image in os.listdir(class_path):
            image_path = os.path.join(class_path, image)

            try:
                # Read file as bytes
                img_bytes = tf.io.read_file(image_path)

                # Decode image (supports jpg, png, gif, etc.)
                img = tf.image.decode_image(
                    img_bytes,
                    channels=3,
                    expand_animations=False
                )

                # Force execution to detect corrupted images
                _ = img.shape

            except Exception as e:
                print(f"Removing corrupted image: {image_path}")
                removed_files.append(image_path)
                tf.io.gfile.remove(image_path)

print("\nSummary:")
if removed_files:
    print("Removed files:")
    for f in removed_files:
        print(f)
else:
    print("No corrupted images found!")
```
